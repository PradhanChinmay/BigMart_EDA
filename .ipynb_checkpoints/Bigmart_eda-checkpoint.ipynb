## Data Preparation and Cleaning

```python
itemWeight_Mean=bigmart_df['Item_Weight'].mean()
itemWeight_Mean
```

```python
bigmart_df['Item_Weight']=bigmart_df['Item_Weight'].fillna(itemWeight_Mean)
```

```python
bigmart_df
```

```python
bigmart_df.info()
```

```python
missing_columns = bigmart_df.columns[bigmart_df.isnull().any()].tolist()
impute_data = bigmart_df[missing_columns]
```

```python
no_missing_data = bigmart_df.drop(missing_columns, axis = 1)
```

```python
knn_imputer = KNNImputer(n_neighbors = 5)
```

```python
imputed_data=pd.DataFrame()
for column in missing_columns:
    if impute_data[column].isnull().any():
        impute_data_numeric = pd.get_dummies(impute_data[column])
        
        imputed_data_numeric = pd.DataFrame(knn_imputer.fit_transform(impute_data_numeric), columns=impute_data_numeric.columns)
        
        imputed_data[column] = imputed_data_numeric.idxmax(axis=1)
    else:
        imputed_data[column] = impute_data[column]

filled_data = pd.concat([no_missing_data, imputed_data], axis=1)
```

```python
new_bigmart_df=filled_data
```

```python
new_bigmart_df
```

```python
columns = new_bigmart_df.columns.tolist()
columns.remove("Outlet_Size")

target_index = columns.index("Outlet_Establishment_Year")
columns.insert(target_index + 1, "Outlet_Size")

new_bigmart_df=new_bigmart_df[columns]
```

```python
new_bigmart_df
```

```python
new_bigmart_df.info()
```

```python
new_bigmart_df=new_bigmart_df.round(2)
```

```python
new_bigmart_df.sample(10)
```

```python
new_bigmart_df["Item_Fat_Content"].unique()
```

```python
new_bigmart_df["Item_Fat_Content"]=new_bigmart_df["Item_Fat_Content"].replace(['low fat','LF'],'Low Fat')
new_bigmart_df["Item_Fat_Content"]=new_bigmart_df["Item_Fat_Content"].replace(['reg'],'Regular')
```

```python
new_bigmart_df.sample(20)
```

## Performing Exploratory Data Analysis

```python
new_bigmart_df.groupby(['Outlet_Establishment_Year', 'Outlet_Identifier'])['Item_Outlet_Sales'].sum().reset_index().sort_values('Item_Outlet_Sales', ascending=True)
```

#### Analyzing the sales grouped based on different outlet types.

```python
df1 = new_bigmart_df.groupby('Outlet_Type')['Item_Outlet_Sales'].sum().reset_index().sort_values('Item_Outlet_Sales', ascending=True)
df1
```

```python
import matplotlib
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#D4E166'
matplotlib.rcParams['font.weight'] = 'bold'

#fff2eb
#EACB01

#DBF406
```

```python
plt.figure(figsize=(12, 7))
num_bars = len(df1.Outlet_Type)
colors = np.linspace(0.4, 0.8, num_bars)
plt.bar(df1.Outlet_Type, df1.Item_Outlet_Sales, width=0.5, color=plt.cm.Greens(colors))
plt.xlabel('Outlet Type', fontsize=12)
plt.ylabel('Sales in millions', fontsize=12)
plt.title('Sales by different Outlet Types', fontsize=14, fontweight="bold");
```

#### Best selling product outlet-wise

```python
df2 = new_bigmart_df.groupby([new_bigmart_df.Outlet_Identifier,new_bigmart_df.Item_Type]).Item_Outlet_Sales.sum().reset_index()
df2
```

```python
max_sale_idx = df2.groupby('Outlet_Identifier')['Item_Outlet_Sales'].idxmax()

# Selecting the rows with the maximum sales for each outlet identifier
max_sale_rows = df2.loc[max_sale_idx]

# extracting the results into a new dataframe
max_sale_df = max_sale_rows[['Outlet_Identifier', 'Item_Type', 'Item_Outlet_Sales']]
```

```python
max_sale_df
```

```python
plt.figure(figsize=(12, 7))
plt.bar(max_sale_df.Outlet_Identifier, max_sale_df.Item_Outlet_Sales, color="Green");
threshold = 400000
ax = plt.gca()
for rect, label in zip(ax.patches, max_sale_df['Item_Type']):
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    if height < threshold:
        ax.text(x + width / 2, y + height , label, ha='center', va='bottom', rotation=90)
    else:
        ax.text(x + width / 2, y + height / 2, label, ha='center', va='center', rotation=90)
plt.xlabel('Outlet', fontsize=12)
plt.ylabel('Sales in millions', fontsize=12)
plt.title('Best selling product outlet-wise', fontsize=14, fontweight="bold");
```

#### Analyzing and plotting the quantity of item sold with respect to the item type.

```python
new_bigmart_df['Item_Quantity_Sold']=new_bigmart_df['Item_Outlet_Sales']/new_bigmart_df['Item_MRP']
new_bigmart_df['Item_Quantity_Sold']=new_bigmart_df['Item_Quantity_Sold'].apply(np.floor)
new_bigmart_df
```

```python
df3=new_bigmart_df.groupby('Item_Type')['Item_Quantity_Sold'].sum().reset_index()
df3 = df3.sort_values('Item_Quantity_Sold', ascending=False)
df3
```

```python
plt.rcParams['axes.labelsize'] = 12
matplotlib.rcParams['figure.facecolor'] = 'w'
plt.figure(figsize=(16,8))
sns.barplot(y=df3.Item_Type, x=df3.Item_Quantity_Sold, palette='Greens_r', orient='h')
plt.title('Quantity of each Item sold', fontsize=14, fontweight="bold");
```
